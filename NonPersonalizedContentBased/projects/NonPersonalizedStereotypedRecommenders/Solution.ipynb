```python
from pprint import pprint
import pandas as pd
import numpy as np
```

# Loading data

```python
data = pd.read_csv('HW1-data.csv')

display(data)
```

# Mean calculation

Calculate the mean rating for each movie, order with the highest rating listed first.

```python
mean_ratings = dict(data.iloc[:,2:].mean())

mean_ratings_sorted = {k: round(v, 3) for k, v in sorted(mean_ratings.items(),
                                                         key=lambda item: item[1],
                                                         reverse=True)}

pprint(mean_ratings_sorted, sort_dicts=False)
```

# Popularity calculation

Count the number of ratings for each movie, order with the most number of ratings first.

```python
rating_count = dict(data.iloc[:,2:].count())

rating_count_sorted = {k: round(v, 3) for k, v in sorted(rating_count.items(),
                                                         key=lambda item: item[1],
                                                         reverse=True)}

pprint(rating_count_sorted, sort_dicts=False)
```

# % of ratings 4+ (liking)

Calculate the percentage of ratings for each movie that are 4 or higher.

```python
percentage_liked = {}

for column in data.columns[2:]:
    
    total = 0
    liked = 0
    
    for rate in data[column]:
        if rate == rate:
            total += 1
        if rate >= 4.0:
            liked += 1
            
    percentage_liked[column] = round(liked / total, 3)
        
percentage_liked_sorted = {k: round(v, 3) for k, v in sorted(percentage_liked.items(),
                                                             key=lambda item: item[1],
                                                             reverse=True)}

pprint(percentage_liked_sorted, sort_dicts=False)
```

# Top movies for someone who has seen 

Calculate movies that most often occur with Toy Story, using the $\frac{x+y}{x}$ method.

In other words, for each movie, calculate the percentage of Toy Story raters who also rated that movie. Order with the highest percentage first.

```python
def association(movie1, movie2):
    total = 0
    sum_both = 0
    for rate1, rate2 in zip(data[movie1], data[movie2]):
        if rate1 == rate1:
            total += 1
        if rate1 == rate1 and rate2 == rate2:
            sum_both += 1
    return round(sum_both / total, 3)
```

```python
movie_to_find_assosiations = '1: Toy Story (1995)'

associations = {}
for movie in data.columns[2:]:
    associations[movie] = association(movie_to_find_assosiations,
                                      movie)
    

associations_sorted = {k: round(v, 3) for k, v in sorted(associations.items(),
                                                         key=lambda item: item[1],
                                                         reverse=True)}

pprint(associations_sorted, sort_dicts=False)
```

# Correlation

Calculate the correlation between the vectors of ratings for Toy Story and each other movie.  

```python
movie_to_find_correlations = '1: Toy Story (1995)'
correlation = dict(data.iloc[:,2:].corr()[movie_to_find_correlations])

correlation_sorted = {k: round(v, 3) for k, v in sorted(correlation.items(),
                                                 key=lambda item: item[1],
                                                 reverse=True)}

pprint(correlation_sorted, sort_dicts=False)
```

# Gender differences

```python
data_male   = data[data["Gender (1 =F, 0=M)"] == 0]
data_female = data[data["Gender (1 =F, 0=M)"] == 1]


rating_count_male = dict(data_male.iloc[:,2:].mean())
rating_count_male_sorted = {k: round(v, 3) for k, v in sorted(rating_count_male.items(),
                                                              key=lambda item: item[1],
                                                              reverse=True)}

male_mean = round(np.average(list(rating_count_male.values())), 5)
print("male:")
pprint(rating_count_male_sorted, sort_dicts=False)
print(f"\naverage male rating: {male_mean}")

print("\n-------------------------\n")
rating_count_female = dict(data_female.iloc[:,2:].mean())
rating_count_female_sorted = {k: round(v, 3) for k, v in sorted(rating_count_female.items(),
                                                              key=lambda item: item[1],
                                                              reverse=True)}
female_mean = round(np.average(list(rating_count_female.values())), 5)
print("female:")
pprint(rating_count_female_sorted, sort_dicts=False)
print(f"\naverage female rating: {female_mean}")

male_female_diff = {}

for movie in rating_count_male.keys():
    male_female_diff[movie] = rating_count_male[movie] - rating_count_female[movie]
        
male_female_diff_sorted = {k: round(v, 3) for k, v in sorted(male_female_diff.items(),
                                                              key=lambda item: item[1],
                                                              reverse=True)}
print("\n-------------------------\n")
print("male - female:")
pprint(male_female_diff_sorted, sort_dicts=False)

female_male_diff = {}

for movie in rating_count_female.keys():
    female_male_diff[movie] = rating_count_female[movie] - rating_count_male[movie]
    
female_male_diff_sorted = {k: round(v, 3) for k, v in sorted(female_male_diff.items(),
                                                              key=lambda item: item[1],
                                                              reverse=True)}
print("\n-------------------------\n")
print("female - male:")
pprint(female_male_diff_sorted, sort_dicts=False)

print("\n-------------------------\n")
print(f"difference between mean ratings: {round(female_mean - male_mean, 3)}")
```
